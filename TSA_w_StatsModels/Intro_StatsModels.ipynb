```python
df = pd.read_csv('macrodata.csv',index_col=0,parse_dates=True)
df.head()
```

```python
df['realgdp'].plot(figsize=(12,5))
```

```python
#Hodrick-Prescott filter -> separate time series into trend and cyclical component
#yt = Ï„t + ct
#Minimizes a quadratic loss function, where lambda is the smoothing parameter
#Rule of thumb for lambda
# 1600 for quarterly data, 6.25 for annual data, 129600 for monthly data
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
gdpcycle, gdptrend = hpfilter(df['realgdp'],lamb=1600)
```

```python
type(gdptrend)
```

```python
gdptrend.plot()
```

```python
df['trend'] = gdptrend
```

```python
df[['trend','realgdp']]['2005-01-01':].plot(figsize=(12,5))
```
