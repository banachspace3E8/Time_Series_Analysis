```python
#Additive models use when trend is more linear and seasonality and trend components seem to be constant over time
#Multipicative models are more appropriate when we are changing at a non-linear rate
```

```python
airline = pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True)
airline = airline.dropna()
airline.head()
```

```python
airline.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,5

result.plot();
```
