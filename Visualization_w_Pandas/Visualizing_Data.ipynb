```python
df1 = pd.read_csv('df1.csv', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2.csv')
df2.head()
```

```python
df1['A'].plot.hist()
```

```python
df1['A'].plot.hist(bins=20,edgecolor='k').autoscale(enable=True,axis='both',tight=True)
```

```python
df1['A'].plot.hist(grid=True)
```

```python
df2.plot.bar(stacked=True)#barh for horizontal plot
```

```python
df2.plot.line(y=['a','b','c'],figsize=(10,4),lw=4)
```

```python
df2.plot.area(stacked=False,alpha=0.4)#alpha for transparency
```

```python
df1.plot.scatter(x='A', y='B',c='C',cmap='coolwarm')#c for colorby
```

```python
df1.plot.scatter(x='A', y='B',s=df1['C']*50,c='D',cmap='coolwarm', alpha=0.3) #S for size
```

```python
df2.plot.box()#.boxplot() also works
```

```python
#Kernel Density Estimation Plot
#For estimating underlying distribution
df2['a'].plot.kde()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.plot.scatter(x='a',y='b')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```
